### Question 5

### Question 7
Analyze the life expectancy dataset (which can be loaded by running the cell below).

Calculate the number of unique countries in the dataset.

Calculate the country and year with the highest life expectancy.

Calculate the year with the highest total expenditure.

Provide your code using pandas.


````python
import seaborn as sns
life_expectancy = sns.load_dataset('healthexp')
````

````python
#Part 1
#Grouping data ponts for the same country together
life_expectancy.groupby('Country').sum() 

#Part 2a
#Grouping data points for the same country
#Calculating the mean value for Life Expectancy for each country
#Sorting the values from highest to lowest
mean_life_by_country = life_expectancy.groupby('Country')['Life_Expectancy'].mean().sort_values(ascending = False)
#Showing the ordered data in the new data frame
mean_life_by_country
#Part 2b
#Grouping data points for the same year
#Calculating the mean value for Life Expectancy for each year
#Sorting the values from highest to lowest
mean_life_by_year = life_expectancy.groupby('Year')['Life_Expectancy'].mean().sort_values(ascending = False)
#Showing the ordered data in the new data frame
mean_life_by_year

#Part 3
#Grouping data points for the same year
#Calculating the mean value for expendture for each year
#Sorting the values from highest to lowest
expenditure_by_year = life_expectancy.groupby('Year')['Spending_USD'].sum().sort_values(ascending = False)
#Showing the ordered data in the new data frame
expenditure_by_year
````

### Question 8
Using the diamonds dataset and the plotting code of your choice, plot the following from the diamonds dataset.

1. Carat vs Price in a scatter plot.
2. A histogram of prices.
3. A boxplot of cut vs price.

The code can be loaded from seaborn in the cell below.

````python
diamonds = sns.load_dataset('diamonds')
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline  
#Part 1

#plotting scatter graph with carats as x axis and diamonds as y axis
plt.scatter(diamonds['carat'], diamonds['price'], color = 'blue')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Carat vs Price')
plt.show()

#Part 2
#plotting histogram for price
plt.hist(diamonds['price'], bins=20, color = 'green')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price of Diamonds')
plt.show()

#Part 3
#plotting boxplot for cut x axis and price y axis
sns.boxplot(x='cut', y='price', data=diamonds)
plt.ylabel('Price')
plt.xlabel('Cut')
plt.title('Price by Cut of Diamonds')
plt.show()
````
